## Define functions

#### calcunion()

Calculate the subsidy scheme and coherent returns for a union 

```python
#calculate the subsidy scheme and coherent returns of a union 
def calcunion(union, fracfield=0.1, fractot=0.01, retmat=retmat, sizemat=sizemat):  
    # restrict matrix
    toflowlist = ['istrax_global']
    for c in union:
        toflowlist.append('istrax_to_'+c)
    ids = ['ctry_code','catid','catname'] 
    rets = retmat[retmat['ctry_code'].isin(union)][ids+toflowlist].copy()

    # add patent counts by field
    rets = pd.merge(rets, sizemat, on=ids)
    # total number of innovations union (this is full-counted, meaning there are double counts)
    innocounttot = rets['innocount'].sum()
    rets['innocountunion'] = innocounttot

    # sum up the destination countries to get returns to union as a whole
    toflowlist = []
    for c in union:
        toflowlist.append('istrax_to_'+c)
    rets['istrax_union'] = rets[toflowlist].sum(axis=1)

    ### Make the union policy (sub-dataframe with all fields that are chosen)
    # sort field-countries by decreasing union returns
    rets = rets.sort_values(by='istrax_union', ascending=False) #Istrax is the marginal return to subsidies in a given field of innovation
    # Make cumulative sum of innocount, to showcase how many innovations are integrated
    rets['cumsum'] = rets['innocount'].cumsum()
    rets['subsidized_field'] = round(rets['innocount'] * fracfield)
    rets['subsidized_total'] = round(rets['subsidized_field'].cumsum(), 0)
    rets['total_to_subsidize'] = round(innocounttot * fractot, 0)
    rets['to_be_subsidized'] = rets['total_to_subsidize'] - rets['subsidized_total']
    rets.loc[rets['to_be_subsidized'] < 0, 'subsidized_field'] = rets['subsidized_field'] + rets['to_be_subsidized']
    rets.loc[rets['to_be_subsidized'] < 0, 'to_be_subsidized'] = 0
    
   # Filter out rows where subsidized_field is greater than 0
    rets = rets[rets['subsidized_field'] > 0]
    unionpol = rets.copy()
    
    # Extract innovation counts per country and store them in a dictionary
    country_innovation_counts = {}
    for country_code in union:
        country_data = rets[rets['ctry_code'] == country_code]
        innovation_count = country_data['subsidized_field'].sum()
        country_innovation_counts[country_code] = innovation_count

    # Calculate weighted sum
    weighted_sum = (unionpol['istrax_union'] * unionpol['subsidized_field']).sum()
    weighted_avg = weighted_sum / unionpol['subsidized_field'].sum()

    return unionpol, weighted_avg, country_innovation_counts
```

#### calc_weights_union()

Calculate the fraction of the total subsidies to a country following a union policy

```python
def calc_weights_union(union, country_innovation_counts): 
    total_innovation_count_union = sum(country_innovation_counts.values())
    weights_union = {}
    for country, count in country_innovation_counts.items():
        if country in union:
            weights_union[country] = count / total_innovation_count_union
    return weights_union 
```

#### calc_weights_individual()

Calculate the fraction of the total subsidies to a country following a national policy

```python
def calc_weights_individual(union):
    
    countrysize = sizemat.groupby('ctry_code')[['innocount']].sum().sort_values('innocount',ascending=False).reset_index()
    unioncountrysize = countrysize[countrysize['ctry_code'].isin(union)].copy()
    unioncountrysize['sum']=unioncountrysize['innocount'].sum()
    unioncountrysize['equalw'] = unioncountrysize['innocount']/unioncountrysize['sum']
    weights_individual = unioncountrysize.set_index('ctry_code')['equalw'].to_dict()
    return weights_individual
    
```

#### gini()

Calculate the Inequality dimension, comparing the wiehts in a union with the weight in a national policy

```python
def calc_gini(weights_individual, weights_union):
    # Create DataFrame from dics.
    colnames = ['country', 'share without union', 'share with union']
    df = pd.DataFrame({'share without union':weights_individual, 'share with union':weights_union})

    # Sort the DataFrame by 'share with union' in ascending order. This assures that the cumulative share with union is always smaller or equal to the share without union
    df = df.sort_values(['share with union','share without union'], ascending=True)

    # Calculate cumulative shares
    df['cumulative share without union'] = df['share without union'].cumsum()
    df['cumulative share with union'] = df['share with union'].cumsum()
    
    # Calculate surfaces under both cumulative shares.
    df['surfwithout'] = df['cumulative share without union']*df['share without union']
    df['surfwith'] = df['cumulative share with union']*df['share without union']

    # Subtract the surface of the share with union from the share without union
    if df['surfwithout'].sum()>0:
        gini =  round((df['surfwithout'].sum() - df['surfwith'].sum())/df['surfwithout'].sum(),4)
    else:
        gini = 0
    return gini
```

#### create_trade_matrix()

The Trade flows are quantified and conglomerated in a country-piars trade matrix

```python
def create_trade_matrix(country_list):
    # Create an empty DataFrame to store the matrix
    matrix = pd.DataFrame(index=country_list, columns=country_list)
    
    directory =  "C:/Users/lucas/Thesis/data/Export"
    total_trade = {}
    iso_file_path = os.path.join(directory, 'ISO.xlsx')
    iso_codes_df = pd.read_excel(iso_file_path)

    for country_code in country_list:
        file_path = os.path.join(directory, f'Export_{country_code}.xlsx')
        if os.path.exists(file_path):
            Export = pd.read_excel(file_path)
            Export['ISO 3'] = Export['ISO 3'].str.upper()
            # Merge with iso_codes_df to convert Alpha-3 to Alpha-2 codes
            df_merged = pd.merge(Export, iso_codes_df[['Alpha-2', 'ISO 3']], how='inner', on='ISO 3')
            total_trade[country_code] = df_merged['Trade Value'].sum()
            filtered_df = df_merged[df_merged['Alpha-2'].isin(country_list)]
    
            # Update the matrix with the export data
            for index, row in filtered_df.iterrows():
                source_country = country_code
                destination_country = row['Alpha-2']
                trade_value = row['Trade Value']
                matrix.loc[source_country, destination_country] = trade_value
    
    # Fill NaN values with 0
    matrix.fillna(0, inplace=True)
    for country in country_list:
        matrix.loc[country] /= total_trade[country]
    for col in matrix.columns:
        if col in matrix.index:
            matrix.loc[col, col] = 1
    return matrix
```

#### cut_matrix()

```python
def cut_matrix(matrix,country_list):
    """
    matrix should have country codes as index and columns. The function will then cut the matrix so we only have those in country_list
    """
    cutmat = matrix.copy()
    cutmat = cutmat.loc[country_list,country_list]
    return cutmat
```

#### genF()

Transform the country-pair matrices to specific union feasibility scores

```python
def genF(similarity_matrix):
    mat = similarity_matrix.copy()
    #Size of matrix is the number of rows of the dataframe times the number of columns:
    NN=mat.shape[0]**2
    #Multiply all elements of the df with (1/NN)
    mat = mat*(1/NN)
    #Sum up all elements (first over rows, then over columns)
    F = mat.sum().sum()
    return F
```

#### generate_country_combinations()

Out of the country list, every possible union is formed

```python
def generate_country_combinations(country_list):
    """
    Generate all possible combinations of countries from the input list.
    """
    country_combinations = []
    for country in range(1, len(country_list) + 1):
        for combination in itertools.combinations(country_list, country):
            country_combinations.append(list(combination))
    return country_combinations
```

```python
#add the country codes individually to the dataframe
def dflist_to_columns(df,listcol,colname="c"):
    adf = df.copy()
    # Expanding the lists into separate columns
    max_length = df[listcol].apply(len).max()  # Find the maximum length of any list
    for i in range(max_length):
        adf[colname+str(i+1)] = df[listcol].apply(lambda x: x[i] if len(x) > i else None)

    adf['size'] = adf[listcol].apply(len)
    return adf

def list_to_dummies(df,listcol):
    adf = df[[listcol]].reset_index().rename(columns={'index':'id'})
    #function to make a row for each list element
    exploded_df = adf.explode(listcol)
    # Creating dummy variables for each country code
    dummies = pd.get_dummies(exploded_df[listcol])
    adf = adf.join(dummies.groupby(exploded_df.index).sum())

    return adf
    
df_processed = dflist_to_columns(df, 'union')
df_processed = pd.concat([df_processed, list_to_dummies(df_processed, 'union')], axis=1)
df_processed
```
